```python
import pickle
import matplotlib.pyplot as plt
```

```python
# this code is to rename the pickle files back to its original names
# because opt.py resets the index back to zero

import os

for file_name in os.listdir("poisoned_data"):
    if file_name.endswith(".p"):
        # Extract the number from the filename
        base_name = os.path.splitext(file_name)[0]
        try:
            num = int(base_name)
            new_num = num + 300
            new_file_name = f"{new_num}.p"
            old_path = os.path.join("poisoned_data", file_name)
            new_path = os.path.join("poisoned_data", new_file_name)
            os.rename(old_path, new_path)
        except ValueError:
            print(f"Filename {file_name} does not contain a valid number.")
```

```python
from PIL import Image
import numpy as np
import os

directory = r"C:\Users\saffr\OneDrive - Nanyang Technological University\NTU\Y4 FYP\fyp_nightshade\poisoned_dataset2"

for file_name in os.listdir("poisoned_data"):
    if file_name.endswith(".p"):
        file_path = os.path.join("poisoned_data", file_name)
        with open(file_path, 'rb') as f:
            data = pickle.load(f)
        print(data)
        image_data = data['img']
        plt.imshow(np.array(image_data))
        txt_data = data['text']
        img_filename = f"{txt_data}.jpg"
        img_path = os.path.join(directory, img_filename)
        img_jpg = Image.fromarray(np.array(image_data))
        img_jpg.save(img_path)
        txt_filename = f"{txt_data}.txt"
        txt_path = os.path.join(directory, txt_filename)
        with open(txt_path, 'w') as txt_file:
            txt_file.write(txt_data)
```
