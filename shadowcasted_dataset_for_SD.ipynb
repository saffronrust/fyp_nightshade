```python
from PIL import Image
import os
import json

# Folder containing the images
input_folder = "image"
output_folder = "nightshade_then_shadowcast"

# Ensure the output folder exists
os.makedirs(output_folder, exist_ok=True)

# Iterate through all files in the input folder
for filename in os.listdir(input_folder):
    if filename.lower().endswith(('.png', '.jpg', '.jpeg')):
        # Load the JSON file
        with open(os.path.join(input_folder, 'cap.json'), 'r') as json_file:
            data = json.load(json_file)

        # Extract the text data for the current image
        print(filename)
        actual_filename = os.path.splitext(filename)[0]
        image_id = actual_filename
        txt_data = next((item['caption'] for item in data['annotations'] if item['image_id'] == image_id), None)
        new_filename = txt_data + ".jpg" 
        print(txt_data)
        print(image_id)
        # Create a text file with the name txt_data and write txt_data into it
        if txt_data:
            txt_file_path = os.path.join(output_folder, f"{txt_data}.txt")
            with open(txt_file_path, 'w') as txt_file:
                txt_file.write(txt_data)
        if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, new_filename)
            
            # Open the image
            with Image.open(input_path) as img:
                # Check if the image size is 336x336
                if img.size == (336, 336):
                    # Resize the image to 512x512
                    upscaled_img = img.resize((512, 512), Image.LANCZOS)
                    # Save the upscaled image
                    upscaled_img.save(output_path)
                else:
                    print(f"Skipping {filename}: Image size is not 336x336.")
```
